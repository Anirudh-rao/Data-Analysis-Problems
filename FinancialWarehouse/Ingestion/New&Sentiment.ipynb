# New & Sentiement Ingestion

In the notebook we will ingest some New and Sentiement data from Alpha Advantage. We will deep dive ingesting some New and Sentiement data across different sectors . For More follow this [link](https://www.alphavantage.co/documentation/)



```python
import os
from dotenv  import load_dotenv
import requests
import pandas as pd
```

## 1. Blockchain :

```python
# Getting API keys
#Load the enviormnet variable for this file
load_dotenv()

#Get env key
AlphaKey = os.getenv('AlphaAdvantage')
```

```python
url = f'https://www.alphavantage.co/query?function=NEWS_SENTIMENT&topics=blockchain&apikey={AlphaKey}'
r = requests.get(url)
data = r.json()

BlockChainDf = pd.DataFrame(data)
BlockChainDf.head(10)
```
